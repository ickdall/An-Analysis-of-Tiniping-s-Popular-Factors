```python
import cv2
from ultralytics import YOLO

# best.pt ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = YOLO('/content/season1_test_1.pt')

# ì²˜ë¦¬í•  ì˜ìƒ íŒŒì¼ ê²½ë¡œ
video_path = '/content/season1_51.mp4'  # ì—¬ê¸°ì— ë¹„ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ì…ë ¥
output_video_path = '/content/final_season1_51.mp4'  # ê²°ê³¼ ë¹„ë””ì˜¤ ì €ì¥ ê²½ë¡œ

# OpenCVë¥¼ ì´ìš©í•´ ë¹„ë””ì˜¤ ì½ê¸°
cap = cv2.VideoCapture(video_path)

# ì›ë³¸ ì˜ìƒì˜ í”„ë ˆì„ ë„ˆë¹„, ë†’ì´, FPS ì •ë³´
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# ê²°ê³¼ ë¹„ë””ì˜¤ ì‘ì„±ì„ ìœ„í•œ VideoWriter ê°ì²´ ìƒì„±
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # ì½”ë± ì„¤ì •
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # ë¹„ë””ì˜¤ì˜ ëì— ë„ë‹¬í•˜ë©´ ì¢…ë£Œ

    # YOLO ëª¨ë¸ë¡œ ê°ì²´ íƒì§€ ìˆ˜í–‰
    results = model(frame)

    # ê²°ê³¼ í”„ë ˆì„ì— ë°•ìŠ¤ì™€ ë¼ë²¨ ê·¸ë¦¬ê¸°
    for result in results:
        boxes = result.boxes  # íƒì§€ëœ ë°•ìŠ¤ ì •ë³´
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]  # ë°•ìŠ¤ ì¢Œí‘œ
            conf = box.conf[0]  # ì‹ ë¢°ë„
            cls = int(box.cls[0])  # í´ë˜ìŠ¤ ì¸ë±ìŠ¤

            # ì‹ ë¢°ë„(conf)ê°€ 0.6 ì´ìƒì¼ ë•Œë§Œ ê²€ì¶œ
            if conf >= 0.6:
                label = f'{model.names[cls]} {conf:.2f}'  # ë¼ë²¨ ë° ì‹ ë¢°ë„

                # ë°•ìŠ¤ ê·¸ë¦¬ê¸°
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # ê²°ê³¼ í”„ë ˆì„ì„ ë¹„ë””ì˜¤ì— ì¶”ê°€
    out.write(frame)

# ë¹„ë””ì˜¤ ê°ì²´ í•´ì œ
cap.release()
out.release()
cv2.destroyAllWindows()

print(f'ê°ì²´ íƒì§€ ê²°ê³¼ ë¹„ë””ì˜¤ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤: {output_video_path}')
```

# í”„ë ˆì„ë³„ class ê°ì§€ count


```python
import os
import cv2
import pandas as pd
from ultralytics import YOLO

# YOLO ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = YOLO('/content/season1_loyal.pt')

# ë¹„ë””ì˜¤ íŒŒì¼ ê²½ë¡œ ì„¤ì •
video_directory = '/content/video/ì‹œì¦Œ1'

# í´ë˜ìŠ¤ ì¹´ìš´íŠ¸ë¥¼ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬ ì´ˆê¸°í™”
class_counts = {}

# ë¹„ë””ì˜¤ ê²½ë¡œì— ìˆëŠ” ëª¨ë“  íŒŒì¼ ì²˜ë¦¬
for video_file in os.listdir(video_directory):
    video_path = os.path.join(video_directory, video_file)

    # OpenCVë¥¼ ì´ìš©í•´ ë¹„ë””ì˜¤ ì½ê¸°
    cap = cv2.VideoCapture(video_path)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # ë¹„ë””ì˜¤ì˜ ëì— ë„ë‹¬í•˜ë©´ ì¢…ë£Œ

        # YOLO ëª¨ë¸ë¡œ ê°ì²´ íƒì§€ ìˆ˜í–‰
        results = model(frame)

        # ê²°ê³¼ì—ì„œ í´ë˜ìŠ¤ ì¹´ìš´íŠ¸ ì¶”ê°€
        for result in results:
            boxes = result.boxes  # íƒì§€ëœ ë°•ìŠ¤ ì •ë³´
            for box in boxes:
                cls = int(box.cls[0])  # í´ë˜ìŠ¤ ì¸ë±ìŠ¤
                class_name = model.names[cls]  # í´ë˜ìŠ¤ ì´ë¦„

                # í´ë˜ìŠ¤ ì¹´ìš´íŠ¸ë¥¼ ë”•ì…”ë„ˆë¦¬ì— ì €ì¥
                if class_name in class_counts:
                    class_counts[class_name] += 1
                else:
                    class_counts[class_name] = 1

    # ë¹„ë””ì˜¤ ê°ì²´ í•´ì œ
    cap.release()

# í´ë˜ìŠ¤ ì¹´ìš´íŠ¸ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
df_class_counts = pd.DataFrame(list(class_counts.items()), columns=['Class', 'Count'])

# ë°ì´í„°í”„ë ˆì„ ì¶œë ¥ ë° ì €ì¥
print(df_class_counts)
df_class_counts.to_csv('/content/class_counts.csv', index=False)

print("ëª¨ë“  ì˜ìƒì—ì„œ íƒì§€ëœ ê°ì²´ì˜ í´ë˜ìŠ¤ë³„ ì´ ì¹´ìš´íŠ¸ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤: /content/class_counts.csv")
```
