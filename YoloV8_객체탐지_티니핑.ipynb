# 객체탐지


```python
import cv2
from ultralytics import YOLO

# best.pt 모델 불러오기
model = YOLO('/content/season1_test_1.pt')

# 처리할 영상 파일 경로
video_path = '/content/season1_51.mp4'  # 여기에 비디오 파일 경로 입력
output_video_path = '/content/final_season1_51.mp4'  # 결과 비디오 저장 경로

# OpenCV를 이용해 비디오 읽기
cap = cv2.VideoCapture(video_path)

# 원본 영상의 프레임 너비, 높이, FPS 정보
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# 결과 비디오 작성을 위한 VideoWriter 객체 생성
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # 코덱 설정
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # 비디오의 끝에 도달하면 종료

    # YOLO 모델로 객체 탐지 수행
    results = model(frame)

    # 결과 프레임에 박스와 라벨 그리기
    for result in results:
        boxes = result.boxes  # 탐지된 박스 정보
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]  # 박스 좌표
            conf = box.conf[0]  # 신뢰도
            cls = int(box.cls[0])  # 클래스 인덱스

            # 신뢰도(conf)가 0.6 이상일 때만 검출
            if conf >= 0.6:
                label = f'{model.names[cls]} {conf:.2f}'  # 라벨 및 신뢰도

                # 박스 그리기
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # 결과 프레임을 비디오에 추가
    out.write(frame)

# 비디오 객체 해제
cap.release()
out.release()
cv2.destroyAllWindows()

print(f'객체 탐지 결과 비디오가 저장되었습니다: {output_video_path}')
```

# 프레임별 class 감지 count


```python
import os
import cv2
import pandas as pd
from ultralytics import YOLO

# YOLO 모델 불러오기
model = YOLO('/content/season1_loyal.pt')

# 비디오 파일 경로 설정
video_directory = '/content/video/시즌1'

# 클래스 카운트를 저장할 딕셔너리 초기화
class_counts = {}

# 비디오 경로에 있는 모든 파일 처리
for video_file in os.listdir(video_directory):
    video_path = os.path.join(video_directory, video_file)

    # OpenCV를 이용해 비디오 읽기
    cap = cv2.VideoCapture(video_path)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # 비디오의 끝에 도달하면 종료

        # YOLO 모델로 객체 탐지 수행
        results = model(frame)

        # 결과에서 클래스 카운트 추가
        for result in results:
            boxes = result.boxes  # 탐지된 박스 정보
            for box in boxes:
                cls = int(box.cls[0])  # 클래스 인덱스
                class_name = model.names[cls]  # 클래스 이름

                # 클래스 카운트를 딕셔너리에 저장
                if class_name in class_counts:
                    class_counts[class_name] += 1
                else:
                    class_counts[class_name] = 1

    # 비디오 객체 해제
    cap.release()

# 클래스 카운트를 데이터프레임으로 변환
df_class_counts = pd.DataFrame(list(class_counts.items()), columns=['Class', 'Count'])

# 데이터프레임 출력 및 저장
print(df_class_counts)
df_class_counts.to_csv('/content/class_counts.csv', index=False)

print("모든 영상에서 탐지된 객체의 클래스별 총 카운트가 저장되었습니다: /content/class_counts.csv")
```
